```python
import os
import natsort


def read_corpus(dir_path):
    files = os.listdir(dir_path)
    files = natsort.natsorted([f for f in files if f.startswith("f")])
    corpus = []
    for file_name in files:
        with open(os.path.join(dir_path, file_name)) as file:
            for line in file:
                corpus.append(line)
    return corpus


corpus = read_corpus("data/")
!
```

```python
# This code was inspired by https://arxiv.org/pdf/1508.07909.pdf, http://ethen8181.github.io/machine-learning/deep_learning/subword/bpe.html
from typing import Dict, List, Tuple
from collections import defaultdict
import re
from operator import itemgetter


class BPETokenizer:
    def __init__(self, vocab_size: int) -> None:
        """
        Args:
            vocab_size (int): size of the vocabulary
        EOW end of word
        UNK unknown symbol. For example, we want to tokenize new text, 
        but tokenizer did not see some symbol in training process
        """
        self.vocab_size = vocab_size
        self.EOW = "</w>"
        self.UNK = "</unk>"

    def get_corpus_repr(self, corpus: List[str]):
        """
        This function converts initial token representation
        For example,
        low low low low low-> 'l o w </w>' : 5, 
        lower lower -> 'l o w e r </w>' : 2,

        Args:
            corpus (List[str]): initial corpus

        Returns:
            _type_: converted corpus
        """        
        corpus_repr = defaultdict(int)
        for line in corpus:
            for word in line.split():
                corpus_repr[" ".join(list(word)) + " " + self.EOW] += 1
        return corpus_repr

    def get_pairs_freq(self, corpus_repr: List[str]) -> Dict[Tuple[str, str], int]:
        """
        This function calculates frequency of each pair of symbols

        Args:
            corpus_repr (List[str]): converted corpus

        Returns:
            Dict[Tuple[str, str], int]: dictonary that contains pair of symbols and its frequency
        """             
        pairs = defaultdict(int)
        for word, freq in corpus_repr.items():
            chars = word.split()
            for i in range(len(chars) - 1):
                pairs[(chars[i], chars[i + 1])] += freq
        return pairs

    def get_pairs(self, word:str) -> List[str]:
        """This function returns all pairs of symbols from the given word

        Args:
            word (str): word :)

        Returns:
            List[str]: all pairs
        """        
        pairs = set()
        prev_char = word[0]
        for ch in word[1:]:
            pairs.add((prev_char, ch))
            prev_char = ch
        return pairs

    def merge_tokens(self, best_pair:str, corpus_repr:Dict[str, int]):
        """
        This function merges tokens

        Args:
            best_pair (str): the most frequent pair
            corpus_repr (Dict[str, int]): converted corpus

        Returns:
            Dict[str, int]: converted corpus
        """        

        new_vocab = defaultdict(int)
        pattern = re.escape(" ".join(best_pair))
        replacement = "".join(best_pair)

        for word_in in corpus_repr:
            word_out = re.sub(pattern, replacement, word_in)
            new_vocab[word_out] = corpus_repr[word_in]

        return new_vocab

    def train(self, corpus: List[str]):
        """
        This function trains bpe

        Args:
            corpus (List[str]): corpus

        Returns:
            set: vocabulary
        """        
        corpus_repr = self.get_corpus_repr(corpus)
        self.vocab = set(" ".join(corpus))
        self.bpe_codes = {}
        i = 0
        while len(self.vocab) < self.vocab_size:
            pairs = self.get_pairs_freq(corpus_repr)
            best_pair = max(pairs, key=pairs.get)
            self.vocab.add("".join(best_pair))
            corpus_repr = self.merge_tokens(best_pair, corpus_repr)
            self.bpe_codes[best_pair] = i
            i += 1
        return self.vocab

    def create_new_word(
        self, word: List[str], pair_to_merge: Tuple[str, str]
    ) -> List[str]:
        """
        This function tokenizes word

        Args:
            word (List[str]): word
            pair_to_merge (Tuple[str, str]): 

        Returns:
            List[str]: 
        """        
        first, second = pair_to_merge
        new_word = []
        i = 0
        while i < len(word):
            try:
                j = word.index(first, i)
                new_word.extend(word[i:j])
                i = j
            except ValueError:
                new_word.extend(word[i:])
                break

            if i < len(word) - 1 and word[i + 1] == second:
                new_word.append(first + second)
                i += 2
            else:
                new_word.append(first)
                i += 1
        return new_word

    def encode(self, original_word: str) -> List[str]:
        """
        This function Encodes word

        Args:
            original_word (str)

        Returns:
            List[str]: tokenized word
        """        
        if len(original_word) == 1:
            return original_word

        word = list(original_word)
        word.append(self.EOW)

        while True:
            pairs = self.get_pairs(word)
            bpe_codes_pairs = [
                (pair, self.bpe_codes[pair]) for pair in pairs if pair in self.bpe_codes
            ]
            if not bpe_codes_pairs:
                break

            pair_to_merge = min(bpe_codes_pairs, key=itemgetter(1))[0]
            word = self.create_new_word(word, pair_to_merge)

        return word

    def tokenize(self, text: str) -> List[str]:
        """
        This function tokenizes text

        Args:
            text (str): text

        Returns:
            List[str]: tokens
        """        
        tokens = []
        for word in text.split():
            tokens.extend(self.encode(word))
        return tokens
```

### Task 2.2. Analysis on Tweets Dataset (10 points)

Train the BPE tokenizer on the tweets dataset. Try to tokenize the tweets with the tokenizer of different `vocab_size`. For example, train the BPE tokenizer with `vocab_size` of [base_vocab_size, 250, 500, 750, 1000]. Plot the dependency of the average length of the tokenized tweet by `vocab_size` to analyze how `vocab_size` affects the length of the tokenized tweet on average. Tell what `vocab_size` is preferrable and why.

```python
import matplotlib.pyplot as plt

vocab_sizes = [250, 500, 750, 1000]
averages = []
for vocab_size in vocab_sizes:
    tokenizer = BPETokenizer(vocab_size)
    tokenizer.train(corpus)
    sum_ = 0
    for i in range(len(corpus)):
        sum_ += len(tokenizer.tokenize(corpus[i]))
    averages.append(sum_ / len(corpus))
```

```python
plt.plot(vocab_sizes, averages)
plt.ylabel('Length of tweet')
plt.xlabel('Vocab size')
plt.show()
```
